```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("powerdata.csv")
print(data)
```

```python
x=data[["Wind_Speed","Blade_Angle","Rotor_Speed"]]
y=data[["Power_Output"]]

model=LinearRegression()
model.fit(x,y)

print(model.coef_)
print(model.intercept_)
print(model.predict([[9,11,100]]))
```
